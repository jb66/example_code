# Update



The aim of this document is to present some of the work done up to this point with the inception v3 algorithm in tensorflow. This work covers training inception v3 to identify band inversion in a band structure. The way this works is that one has to create multiple categories of images to train the algorithm. For example, if one wants the algorithm to identify flowers, one would need multiple categories of flowers - rose, daisy, dandelion, etc. Each category has a folder containing images representative of the category. The code points to each folder. In this particular case, since we are concerned with band inversion, then band inverion was one category. A separate category called "test" was created. This category had images that were NOT representative of band inversion. This other category was needed as the code needs at least two categories to work. Caution was taken to ensure that this other category did not contain features of a band structure. The reason for this is because the code is not yet set up to identify multiple categories in one image simultaneously. It can only identify them one at a time. For it to identify multiple categories in an image simultaneously, the current code needs to be modified appropriately. This will be done once there is confidence that the code correctly identifies all categories of interest one at a time. 

In order to test the code in identifying categories (band inversion in this case) one at a time, a few band structure images were downloaded and images of band inversion from each case were used to train the algorithm.
